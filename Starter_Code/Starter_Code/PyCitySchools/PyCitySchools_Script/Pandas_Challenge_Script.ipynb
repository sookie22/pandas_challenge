```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
school_data_to_load = Path('C:\\Users\\sooki\\OneDrive\\Desktop\\Bootcamp Classwork Projects\\Week 4 - Pandas and Jupyter\\Pandas_Challenge\\pandas_challenge\\Starter_Code\\Starter_Code\\PyCitySchools\\Resources\\schools_complete.csv')
student_data_to_load = Path('C:\\Users\\sooki\\OneDrive\\Desktop\\Bootcamp Classwork Projects\\Week 4 - Pandas and Jupyter\\Pandas_Challenge\\pandas_challenge\\Starter_Code\\Starter_Code\\PyCitySchools\\Resources\\students_complete.csv')

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# Local Government Area (LGA) Summary

Perform the necessary calculations and then create a high-level snapshot of the local government area's key metrics in a DataFrame. 

Include the following:

1) Total number of unique schools
2) Total students
3) Total budget
4) Average maths score
5) Average reading score
6) % passing maths (the percentage of students who passed maths)
7) % passing reading (the percentage of students who passed reading)
8) % overall passing (the percentage of students who passed maths AND reading)

Note: A passing grade is 50 or higher.

```python
# 1) Total number of unique schools

unique_schools = school_data_complete["school_name"].unique()
total_unique_school_number = len(unique_schools)
print("The total number of unique schools is ", total_unique_school_number)
```

```python
# 2) Total students

total_students = len(school_data_complete["Student ID"])
print("The total number of students is", total_students)
```

```python
# 3) Total budget

# Get unique school names
unique_schools = school_data_complete["school_name"].unique()

# Set a variable to store the total budget
total_budget = 0

# Iterate through unique school names and calculate budgets
for school in unique_schools:
    budget = school_data_complete[school_data_complete["school_name"] == school]['budget'].iloc[0]
    total_budget += budget

# Print the sum of budgets for all unique schools
print("Total budget for all unique schools:", total_budget)
```

```python
# 4) Average maths score

avg_maths_score = "{:.3f}".format(sum(school_data_complete["maths_score"] / len(school_data_complete["Student ID"])))
print("The average maths score is ", avg_maths_score)
```

```python
# 5) Average reading score

avg_reading_score = "{:.3f}".format(sum(school_data_complete["reading_score"] / len(school_data_complete["Student ID"])))
print("The average reading score is ", avg_reading_score)
```

```python
# 6) % passing maths (the percentage of students who passed maths)
# Note: A passing grade is 50 or higher.

passing_maths = (school_data_complete['maths_score'] >= 50).sum()
total_students = len(school_data_complete["Student ID"])

passing_maths_percentage = "{:.3f}%".format(passing_maths / total_students * 100)
print("The percentage of students who passed maths is", passing_maths_percentage)
```

```python
# 7)  % passing reading (the percentage of students who passed reading)
# Note: A passing grade is 50 or higher.

passing_reading = (school_data_complete['reading_score'] >= 50).sum()
total_students = len(school_data_complete["Student ID"])

passing_reading_percentage = "{:.3f}%".format(passing_reading / total_students * 100)
print("The percentage of students who passed reading is", passing_reading_percentage)
```

```python
# 8) % overall passing (the percentage of students who passed maths AND reading)
# Note: A passing grade is 50 or higher.

passing_both = ((school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)).sum()
total_students = len(school_data_complete["Student ID"])

passing_both_percentage = "{:.3f}%".format(passing_both / total_students * 100)
print("The percentage of students who passed both maths and reading is", passing_both_percentage)
```

```python
data = {"Total Schools": [total_unique_school_number],
        "Total Students": [total_students],
        "Total Budget": ["${:,.2f}".format(total_budget)],
        "Average Maths Score": [avg_maths_score],
        "Average Reading Score": [avg_reading_score],
        "% Passing Maths": [passing_maths_percentage],
        "% Passing Reading": [passing_reading_percentage],
        "% Overall Passing": [passing_both_percentage]}

LGA_summary_df = pd.DataFrame(data)
LGA_summary_df
```

## School Summary

Perform the necessary calculations and then create a DataFrame that summarises key metrics about each school.

Include the following:

1) School name
2) School type
3) Total students
4) Total school budget
5) Per student budget
6) Average maths score
7) Average reading score
8) % passing maths (the percentage of students who passed maths)
9) % passing reading (the percentage of students who passed reading)
10) % overall passing (the percentage of students who passed maths AND reading)

```python
# Extract the array of unique school names from the "school_name" column of the school_data_complete data
unique_schools = school_data_complete["school_name"].unique()

# Create an empty list to store the summary information for each school as per assignment 
school_summary = []

# Loop through each unique school name
for school in unique_schools:
    
    # School name 
    # Filter the school_data_complete data
    # Include only rows where "school_name" matches a specific school name
    school_data = school_data_complete[school_data_complete["school_name"] == school]
    
    # School type
    # Retrieve the value in the first row of the "type" column for the specific school 
    school_type = school_data["type"].iloc[0]
    
    # Total number of student is the count of each school data
    total_students = len(school_data)
    
    # Total school budget 
    # Retrieve the value in the first row of the "budget" column for the specific school 
    total_budget = school_data["budget"].iloc[0]
    formatted_total_budget = "${:,.2f}".format(total_budget)
    
    # 5) Per student budget 
    per_student_budget = total_budget / total_students
    formatted_per_student_budget = "${:,.2f}".format(per_student_budget)
    
    # 6) and 7) Average score for math and reading
    avg_maths_score = school_data["maths_score"].mean()
    formatted_avg_maths_score = "{:.3f}".format(avg_maths_score)
    
    avg_reading_score = school_data["reading_score"].mean()
    formatted_avg_reading_score = "{:.3f}".format(avg_reading_score)
    
    # 8) % passing maths (the percentage of students who passed maths)
    passing_maths = len(school_data[school_data["maths_score"] >= 50])
    passing_maths_percentage = (passing_maths / total_students) * 100
    formatted_passing_maths_percentage = "{:.3f}".format(passing_maths_percentage)
    
    # 9) % passing reading (the percentage of students who passed reading)
    passing_reading = len(school_data[school_data["reading_score"] >= 50])
    passing_reading_percentage = (passing_reading / total_students) * 100
    formatted_passing_reading_percentage = "{:.3f}".format(passing_reading_percentage)
    
    # 10) % overall passing (the percentage of students who passed maths AND reading)
    passing_both = len(school_data[(school_data["maths_score"] >= 50) & (school_data["reading_score"] >= 50)])
    passing_both_percentage = (passing_both / total_students) * 100
    formatted_passing_both_percentage = "{:.3f}".format(passing_both_percentage)

    
    school_summary.append({
        "School Name": school,
        "School Type": school_type,
        "Total Students": total_students,
        "Total School Budget": formatted_total_budget,
        "Per Student Budget": formatted_per_student_budget,
        "Average Maths Score": formatted_avg_maths_score,
        "Average Reading Score": formatted_avg_reading_score,
        "% Passing Maths": formatted_passing_maths_percentage,
        "% Passing Reading": formatted_passing_reading_percentage,
        "% Overall Passing": formatted_passing_both_percentage
    })

school_summary_df = pd.DataFrame(school_summary)
school_summary_df
```

Include the following metrics in the table:

- % overall passing (the percentage of students who passed maths AND reading)




```python
# Define the spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["< 585", "585-630", "630-645", "645-680"]

# Create a new column in the school_summary_df DataFrame to store the spending range for each school
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=labels)

# Convert the columns to numeric data types
school_summary_df["Average Maths Score"] = pd.to_numeric(school_summary_df["Average Maths Score"], errors='coerce')
school_summary_df["Average Reading Score"] = pd.to_numeric(school_summary_df["Average Reading Score"], errors='coerce')
school_summary_df["% Passing Maths"] = pd.to_numeric(school_summary_df["% Passing Maths"], errors='coerce')
school_summary_df["% Passing Reading"] = pd.to_numeric(school_summary_df["% Passing Reading"], errors='coerce')
school_summary_df["% Overall Passing"] = pd.to_numeric(school_summary_df["% Overall Passing"], errors='coerce')

# Calculate mean scores per spending range
spending_maths_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

# Create a DataFrame to store the spending summary
spending_summary = pd.DataFrame({"Average Maths Score": spending_maths_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Maths": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing_spending})

# Format the values to only have 2 decimal points:
spending_summary = spending_summary.round(2)

spending_summary
```

## Scores by School Size

Use the following code to bin the per_school_summary.

- labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

Use pd.cut on the "Total Students" column of the per_school_summary DataFrame.

Create a DataFrame called size_summary that breaks down school performance based on school size (small, medium, or large).

```python
# Define the size bins and labels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Create a new column in the school_summary_df DataFrame to store the school size range
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], size_bins, labels=labels)

# Calculate mean scores per school size range
size_math_scores = school_summary_df.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = school_summary_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_summary_df.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = school_summary_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_summary_df.groupby(["School Size"])["% Overall Passing"].mean()

# Create the size_summary DataFrame
size_summary = pd.DataFrame({"Average Maths Score": size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Maths": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing": size_overall_passing})

size_summary
```
